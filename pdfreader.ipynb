```python
from PyPDF2 import PdfReader
```

```python
# provide the path of  pdf file/files.
pdfreader = PdfReader('demo/facultymanual.pdf')
```

```python
# read text from pdf
raw_text = ''
for i, page in enumerate(pdfreader.pages):
    content = page.extract_text()
    if content:
        raw_text += content
```

```python
raw_text
```

```python
f = open("demo/facultymanual.txt", "a")
f.write(raw_text)
f.close()
```
